```python
# import important libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

```python
# upload image
pic = Image.open('/00-puppy.jpg')
```

```python
# check type then convert to arry
print(type(pic))
arr_pic = np.asarray(pic)
print(arr_pic.shape)
# show array immage by 3 channel
plt.imshow(arr_pic)
```

```python
# red channel color by defult cmap system
pic_red = arr_pic[:,:,0]
plt.imshow(pic_red)
```

```python
# red channel color by grey cmap system 0 close is black no red color.
plt.imshow(pic_red,cmap='grey')
```

```python
# Green channel color by grey cmap system 0 close is black no green color.
pic_grn = arr_pic[:,:,1]
plt.imshow(pic_grn, cmap='gray')
```

```python
# Blue channel color by grey cmap system 0 close is black no blue color.
pic_blu = arr_pic[:,:,2]
plt.imshow(pic_blu, cmap='gray')
```

```python
# copy new image from pic sourse
arr_image_copy = np.copy(arr_pic)
```

```python
# convert green (G) channel color to 0 arry without change R B. pic after remove green.
arr_image_copy[:,:,1] = 0
plt.imshow(arr_image_copy)
```

```python
# convert Blue (B) channel color to 0 arry without change R  pic after remove blue and green channels.
# we keept red channel just
arr_image_copy[:,:,2] = 0
plt.imshow(arr_image_copy)
```

```python
# Plot RGB distributions
plt.hist(pic_red.ravel(), bins=256, color='red', alpha=0.5, label='Red')
plt.hist(pic_grn.ravel(), bins=256, color='green', alpha=0.5, label='Green')
plt.hist(pic_blu.ravel(), bins=256, color='blue', alpha=0.5, label='Blue')
plt.title("RGB Color Distribution")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.legend()
plt.show()

```
